## บ้าน

### สร้างตาราง

```python
est2_home['ประเมินโดย2']  = est2_home['post id'].map(est2_home.set_index('post id')['ประเมินราคาซื้อ-ขาย'])
```

```python
est2_home['ประเมินโดย1'] = est2_home['post id'].map(est1_home.set_index('post id')['ประเมินราคาซื้อ-ขาย'])
est2_home.head()
```

```python
home_mape = est2_home[['post id','ราคาทรัพย์','ประเมินโดย1','ประเมินโดย2']]
home_mape.isnull().any()
home_mape = home_mape.dropna(subset= ['ประเมินโดย2'])
```

```python
home_mape
```

### MAPE

```python
from sklearn.metrics import mean_absolute_percentage_error #as MAPE
```

```python
a = home_mape['ราคาทรัพย์']
a_pred_1 = home_mape['ประเมินโดย1']
a_pred_2 = home_mape['ประเมินโดย2']
```

```python
mean_absolute_percentage_error(a,a_pred_1 )
print(f'ค่า MAPE ระหว่าง ราคาทรัพย์(บ้าน) และ ราคาของผู้ประเมินคนที่1 เท่ากับ {mean_absolute_percentage_error(a,a_pred_1)*100:.2f}%')
```

```python
mean_absolute_percentage_error(a,a_pred_2)
print(f'ค่า MAPE ระหว่าง ราคาทรัพย์(บ้าน) และ ราคาของผู้ประเมินคนที่2 เท่ากับ {mean_absolute_percentage_error(a,a_pred_2)*100:.2f}%')
```

```python
mean_absolute_percentage_error(a_pred_1,a_pred_2 )
print(f'ค่า MAPE ระหว่าง  ราคาของผู้ประเมินคนที่1 และ ราคาของผู้ประเมินคนที่2 เท่ากับ {mean_absolute_percentage_error(a_pred_1,a_pred_2)*100:.2f}%')
```

## ค่าMAPEก่อนแก้ไขข้อมูล

### บ้าน

```python
#อ่านข้อมูลจาก google sheet 
link = 'https://docs.google.com/spreadsheets/d/1a6NId_-N7tFnAxva721mT0ktVUwaTKoB0BkYvy2S7Ao/export?format=xlsx'
home = pd.read_excel(link,sheet_name='home')
home
```

#### mape

```python
est1_home = home[home['ID ผู้ประเมิน']==1]
est1_home = est1_home.dropna(subset=['ราคาทรัพย์','ประเมินราคาซื้อ-ขาย'])
est2_home = home [home ['ID ผู้ประเมิน']==2]
est2_home = est2_home.dropna(subset=['ราคาทรัพย์','ประเมินราคาซื้อ-ขาย'])
est1_home_mape = mean_absolute_percentage_error(est1_home['ราคาทรัพย์'],pd.to_numeric(est1_home['ประเมินราคาซื้อ-ขาย']))
est2_home_mape = mean_absolute_percentage_error(est2_home['ราคาทรัพย์'],pd.to_numeric(est2_home['ประเมินราคาซื้อ-ขาย']))
print(f'ความแตกต่างระหว่างราคาประกาศขายและราคาประเมินของ\nบ้าน\nนักประเมิน คนที่ 1 = {est1_home_mape*100:.2f}%\nนักประเมิน คนที่ 2 = {est2_home_mape*100:.2f}%')
est2_home['ประเมิน1'] = est2_home['post id'].map(est1_home.set_index('post id')['ประเมินราคาซื้อ-ขาย'])
est2_home_new = est2_home.dropna(subset=['ประเมินราคาซื้อ-ขาย','ประเมิน1','ราคาทรัพย์'])
est2_home_new = est2_home_new[['post id','ราคาทรัพย์','ประเมินราคาซื้อ-ขาย','ประเมิน1']]
one2two = mean_absolute_percentage_error(est2_home_new['ประเมินราคาซื้อ-ขาย'],est2_home_new['ประเมิน1'])
two2one = mean_absolute_percentage_error(est2_home_new['ประเมิน1'],est2_home_new['ประเมินราคาซื้อ-ขาย'])
print(f'ความผิดพลาดระหว่างนักประเมินทั้ง {((one2two)+(two2one))*50:.2f}%')
print(f'จำนวนทรัพย์ทั้งหมด {len(est2_home_new)} ทรัพย์')
```
