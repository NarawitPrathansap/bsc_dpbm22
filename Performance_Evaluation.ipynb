```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.datasets import load_iris

model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import make_scorer, precision_score, recall_score

# Load the iris dataset
iris = load_iris()

# Create the decision tree classifier
clf = DecisionTreeClassifier()

# Define the 10-fold cross-validation procedure
cv = KFold(n_splits=10, shuffle=True, random_state=1)

# Define the precision and recall scorers
precision_scorer = make_scorer(precision_score, average='macro')
recall_scorer = make_scorer(recall_score, average='macro')

# Evaluate the model using cross-validation with precision and recall
precision_scores = cross_val_score(clf, iris.data, iris.target, cv=cv, scoring=precision_scorer)
recall_scores = cross_val_score(clf, iris.data, iris.target, cv=cv, scoring=recall_scorer)

# Output the precision and recall scores for each fold
for i, (train_indexes, test_indexes) in enumerate(cv.split(iris.data)):
    print("Fold", i+1)
    print("Training samples:", len(train_indexes))
    print("Test samples:", len(test_indexes))
    print("Precision:", precision_scores[i])
    print("Recall:", recall_scores[i])
    print("------------------------------")
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
test_fold = 1 
```

```python
test_data = shuffled_df[shuffled_df['fold']==test_fold]
training_data = shuffled_df[shuffled_df['fold']!=test_fold]
```

```python
X_test = test_data.iloc[:,:-2]
y_test = test_data.iloc[:,-2]
X_train = training_data.iloc[:,:-2]
y_train = training_data.iloc[:,-2]
```

```python
model = DecisionTreeClassifier(max_depth=2)

model.fit(X_train,y_train)

y_predict = model.predict(X_test)
```

```python
acc = accuracy_score(y_test,y_predict)
print(f'accuracy of fold {test_fold} = {acc}')
```

```python
acc_fold=[]
for test_fold in range(1,11):
    test_data = shuffled_df[shuffled_df['fold']==test_fold]
    training_data = shuffled_df[shuffled_df['fold']!=test_fold]
    X_test = test_data.iloc[:,:-2]
    y_test = test_data.iloc[:,-2]
    X_train = training_data.iloc[:,:-2]
    y_train = training_data.iloc[:,-2]
    
    model = DecisionTreeClassifier(max_depth=2)
    model.fit(X_train,y_train)
    y_predict = model.predict(X_test)

    acc = accuracy_score(y_test,y_predict)
    print(f'accuracy of fold {test_fold} = {acc}')
    acc_fold.append(acc)

print(f'\naverage accuracy = {np.mean(acc)}')
```
