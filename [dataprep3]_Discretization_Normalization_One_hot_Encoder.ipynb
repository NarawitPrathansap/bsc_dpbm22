```python
data_features_d = data_features[data_features.eq('-').any(1)==False] #หาค่าที่เราต้องการ ใส่Falseเพื่อเอาค่าที่ไม่มี -  
data_features_d = data_features_d[data_features_d.eq('T').any(1)==False] #หาค่าที่เราต้องการ ใส่Falseเพื่อเอาค่าที่ไม่มี T
data_features_d
```

```python
data_features_d.describe()
```

```python
data_features_d['ความเร็วลม'] = data_features_d['ความเร็วลม'].astype(float)
data_features_d['อุณหภูมิต่ำสุด '] = data_features_d['อุณหภูมิต่ำสุด'].astype(float)
data_features_d['จำนวนเมฆ'] = data_features_d['จำนวนเมฆ'].astype(float)
data_features_d['ปริมาณฝน'] = data_features_d['ปริมาณฝน'].astype(float)
```

#StandardScaler

```python
#define บอกว่าวิธีการที่เราจะทำคืออะไร () เป็นพารามิเตอร์ที่เราใส่ได้
scaler = preprocessing.StandardScaler()
#train สร้างโมเดลจาก data 
scaler.fit_transform(data_features_d)
#predict-transform
data_features_d_s = scaler.transform(data_features_d)
```

```python
data_features_d_s
```

```python
plt.hist(data_features_d_s[:,0])
```

```python
plt.subplot(1,2,1)
plt.hist(data_features_d['ความกด']);

plt.subplot(1,2,2)
plt.hist(data_features_d_s[:,0]);
```

```python
plt.subplot(1,2,1)
plt.hist(data_features_d['ปริมาณฝน']);

plt.subplot(1,2,2)
plt.hist(data_features_d_s[:,6]);
```

### Min-max normalization

```python
#define บอกว่าวิธีการที่เราจะทำคืออะไร () เป็นพารามิเตอร์ที่เราใส่ได้
#Scale ปรับขนาดให้อยู่ในมาตรฐาน
scaler = preprocessing.MinMaxScaler()
#train สร้างโมเดลจาก data 
scaler.fit_transform(data_features_d)
#predict-transform
data_features_d_norm = scaler.transform(data_features_d)
```

```python
plt.subplot(1,3,1)
plt.hist(data_features_d['ความกด']);
plt.subplot(1,3,2)
plt.hist(data_features_d_s[:,0]);
plt.subplot(1,3,3)
plt.hist(data_features_d_norm[:,0]);
```

### แปลงค่า sell_price

```python
estate_data[(estate_data['sell_price']>=5000000) & (estate_data['sell_price']<=10000000)]
```

```python
estate_data[(estate_data['sell_price']>=5000000) & (estate_data['sell_price']<=10000000)]
```
